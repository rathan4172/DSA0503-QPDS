2.Write a Pandas program to display the IDs of employees who have done two or more jobs in the past.

```python
import matplotlib.pyplot as plt
import numpy as np

men = [22, 30, 35, 35, 26]
women = [25, 32, 30, 35, 29]

x = np.arange(len(men))
width = 0.35

plt.bar(x - width/2, men, width, label='Men')
plt.bar(x + width/2, women, width, label='Women')

plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

men = [22, 30, 35, 35, 26]
women = [25, 32, 30, 35, 29]
men_sd = [4, 3, 4, 1, 5]
women_sd = [3, 5, 2, 3, 3]

x = np.arange(len(men))

plt.bar(x, men, yerr=men_sd)
plt.bar(x, women, bottom=men, yerr=women_sd)

plt.show()
```
